```python
# Import dependencies
import pandas as pd
```

```python
# Store csv files into variables
MALARIA_RATE = "Resources/Estimated_malaria_incidence_2000_to_2021.csv"
MORTALITY_RATE = "Resources/Estimated_malaria_mortality_rate_2000_to_2010.csv"
MALARIA_CASES = "Resources/Estimated_number_of_malaria_cases_2000_to_2021.csv"
MALARIA_DEATHS = "Resources/Estimated_number_of_malaria_deaths_2000_to_2021.csv"
```

```python
# Create dataframes
malaria_rate_df = pd.read_csv(MALARIA_RATE)
malaria_rate_df.head()
```

```python
# Create dataframes
mortality_rate_df = pd.read_csv(MORTALITY_RATE)
mortality_rate_df.head()
```

```python
# Create dataframes
malaria_cases_df = pd.read_csv(MALARIA_CASES)
malaria_cases_df.head()
```

```python
# Create dataframes
malaria_deaths_df = pd.read_csv(MALARIA_DEATHS)
malaria_deaths_df.head()
```

```python
malaria_deaths_df.columns
```

```python
malaria_deaths_df = malaria_deaths_df[["ParentLocation", "Location", "Period", "FactValueNumeric"]]
malaria_deaths_df.head()
```

```python
malaria_rate_df = malaria_rate_df[["ParentLocation", "Location", "Period", "FactValueNumeric"]]
malaria_rate_df.head()
```

```python
malaria_cases_df = malaria_cases_df[["ParentLocation", "Location", "Period", "FactValueNumeric"]]
malaria_cases_df.head()
```

```python
mortality_rate_df = mortality_rate_df[["ParentLocation", "Location", "Period", "FactValueNumeric"]]
mortality_rate_df.head()
```

```python
malaria_cases_df = malaria_cases_df.rename(columns = {"ParentLocation": "Region", "Location": "Country",
                                                      "Period": "Year", "FactValueNumeric": "Estimated Number of Cases"})
malaria_deaths_df = malaria_deaths_df.rename(columns = {"ParentLocation": "Region", "Location": "Country",
                                                      "Period": "Year", "FactValueNumeric": "Estimated Number of Deaths"})
malaria_rate_df = malaria_rate_df.rename(columns = {"ParentLocation": "Region", "Location": "Country",
                                                      "Period": "Year", "FactValueNumeric": "Estimated Incidence (Per 1,000 Population at Risk)"})
mortality_rate_df = mortality_rate_df.rename(columns = {"ParentLocation": "Region", "Location": "Country",
                                                      "Period": "Year", "FactValueNumeric": "Estimated Mortality Rate (Per 100,000 Population)"})
```
