```python
# Import dependencies
import pandas as pd
from config import api_key
import requests
import json
```

```python
# Import csv files as dataframes
cases_df = pd.read_csv("Output/malaria_cases.csv")
deaths_df = pd.read_csv("Output/malaria_deaths.csv")
rates_df = pd.read_csv("Output/malaria_rate.csv")
mortality_df = pd.read_csv("Output/mortality_rate.csv")
```

```python
# Create a function to get a list of countries that are in each individual table
def unique_countries(dataframe):
    global countries
    countries = dataframe["Country"].unique()
```

```python
# Run the function on each table
unique_countries(cases_df)
cases_countries = countries
cases_countries
```

```python
# Run the function on each table
unique_countries(deaths_df)
deaths_countries = countries
deaths_countries
```

```python
# Run the function on each table
unique_countries(rates_df)
rates_countries = countries
rates_countries
```

```python
# Run the function on each table
unique_countries(mortality_df)
mortality_countries = countries
mortality_countries
```

```python
# Combine all four lists and drop duplicates
countries_list = list(set(cases_countries) | set(deaths_countries) | set(rates_countries) | set(mortality_countries))
print(countries_list)
```

```python
# Create a new dataframe that will have each country and will be populated with latitude and longitude
countries_df = pd.DataFrame({"Country": countries_list, "Latitude": "", "Longitude": ""})
countries_df
```

```python
# Use a geoapify api call to get the coordinates for each country in the dataframe
for index, row in countries_df.iterrows():
        # Loop through zip codes
        country = countries_df.loc[index, "Country"]
        # define target url
        target_url = f"https://api.geoapify.com/v1/geocode/search?country={country}&apiKey={api_key}"
        # make API request
        response = requests.get(target_url).json()
        try:
            countries_df.loc[index, "Latitude"] = response["features"][0]["properties"]["lat"]
            countries_df.loc[index, "Longitude"] = response["features"][0]["properties"]["lon"]
            print(f"{country} found")
        except:
            countries_df.loc[index, "Latitude"] = "Not Found"
            countries_df.loc[index, "Longitude"] = "Not Found"
            print(f"{country} NOT found")
```

```python
# Check the dataframe
countries_df
```

```python
# Find out which countries have an unknown longitude
lon_not_found = countries_df.loc[countries_df["Longitude"] == "Not Found"]
lon_not_found
```

```python
# Populate these countries with coordinates found through google search
countries_df.loc[countries_df["Country"] == "Mayotte", "Latitude"] = -12.8275
countries_df.loc[countries_df["Country"] == "Mayotte", "Longitude"] = 45.1662

countries_df.loc[countries_df["Country"] == "French Guiana", "Latitude"] = 3.9339
countries_df.loc[countries_df["Country"] == "French Guiana", "Longitude"] = -53.1258
```
